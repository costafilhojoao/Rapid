# VAR: Diagnóstico

## VAR(2)

Vamos trabalhar com o primeiro VAR estimado no notebook da aula passada.

```r
install.packages("vars")
```

```r
library(vars)
```

### Teste de autocorrelação serial

```r
serial_test <- vars::serial.test( var, lags.pt = 12, type = "BG")

print(serial_test)
```

### Teste de heterocedasticidade condicional (ARCH-LM)

```r
arch_test <- vars::arch.test( var, lags.multi = 12, multivariate.only = TRUE)

print(arch_test)
```

### Testes de normalidade

```r
normal_test <- vars::normality.test( var, multivariate.only = TRUE)

print( normal_test )
```

### Estabilidade e quebra estrutural

```r
options(repr.plot.width = 14, repr.plot.height = 14)

stability_test <- vars::stability( var, type = "OLS-CUSUM") # Ploberger and Krämer (1992)

plot(stability_test)
```

```r
strucchange::sctest( var )
```
