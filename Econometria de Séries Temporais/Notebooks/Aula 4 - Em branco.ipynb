## Simulação de processos estocásticos

Com base no exemplo do primeiro capítulo de Bueno (2012), trabalhemos com quatro processos estocásticos:

* $y_t = 0.5 t + ɛ_t$
* $y_t = 0.5 t + y_{t-1} + ɛ_t$
* $y_t = y_{t-1} + ɛ_t$
* $y_t = 0.5 + 0.8 y_{t-1} + ɛ_t$

Vamos assumir que $ɛ ∼ N(0, σ^2)$ com $σ^2 = 0.5^{2}$.

Primeiro, vamos gerar uma sequência de 100 choques aleatórios:

```r
set.seed(1301)     # Seed para obtermos os mesmos números pseudo-aleatórios

e <- rnorm( 100, mean = 0, sd = 0.5 )
```

Transforme "e" em uma série de tempo trimestral com início em janeiro de 1998:

Faça um gráfico da variável "e" ao longo do tempo:

Vamos criar a série da tendência para o mesmo número de observações. Para isso, vamos usar a função 'seq()':

```r
time = seq( from = 1, to = 100, by = 1 )
```

Agora, vamos simular o processo $y_t = 0.5 t + ɛ_t$. Chame o objeto de y1:

Crie também as simulações para

$y_t = 0.5 t + y_{t-1} + ɛ_t$, \\
$y_t = y_{t-1} + ɛ_t$ e \\
 $y_t = 0.5 + 0.8 y_{t-1} ɛ_t$.

Consindere $y_0$ igual para todos:

```r
y2 <- rep( NA, length(y1) )

y2[1] <- y1[1]

for (t in 2:length(y2)) {
  y2[t] <- 0.5 * time[t] + y2[t-1] + e[t]
}
```

Coloque y1, y2, y3 e y4 em um mesmo dataframe, junto à uma variável que represente o trimestre:

```r
data = data.frame( y1, y2, y3, y4, quarter = time( y1) )
```

Faça quatro gráficos (p1, p2, p3 e p4), um para cada série temporal (y1, y2, y3 e y4):

```r
library(ggplot2)
```
