## Teste Ljung-Box

$\mathcal{H}_0: \text{As primeiras k autocorrelações são iguais à zero}$ \\
$\mathcal{H}_a: \text{Ao menos uma das primeiras k autocorrelações é diferente de zero}$ \\

```r
Box.test( resids,
          lag = 1,
          type = c( "Ljung-Box" ),
          fitdf = 0  # número de graus de liberdade
          )
```
