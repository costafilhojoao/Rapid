## Filtro HP

Extraia a tendência do IBC-Br com o filtro HP:

```r
install.packages("mFilter")
```

```r
library(mFilter)
```

```r
hp <- mFilter( log(ibc) , filter="HP", freq = 14400)
```

```r
hptrend <- hp$trend
```

Faça um gráfico que contenha duas curvas: o IBC-Br e o exponencial da tendência estimada pelo filtro HP:

Obtenha o componente cíclico com o filtro HP:

```r
hpcycle <- hp$cycle * 100
```

## Filtro Hamilton (2018)

Extraia a tendência do IBC-Br com o filtro proposto por Hamilton (2018):

```r
install.packages("neverhpfilter")
```

```r
library(neverhpfilter)
```

```r
y <- as.xts( log(ibc) )
dimnames(y) <- list(NULL, "log_ibc")
```

```r
hamilton <- yth_filter(y, h = 8, p = 4)
```

```r
hamiltontrend <- hamilton$log_ibc.trend
```

Faça um gráfico que contenha duas curvas: o IBC-Br e o exponencial da tendência estimada pelo filtro proposto por Hamilton (2018):

Obtenha o componente cíclico com o filtro proposto por Hamilton (2018):

```r
hamiltoncycle <- hamilton$log_ibc.cycle * 100
```

Faça um gráfico dele ao longo do tempo:

## Boosted HP

Extraia a tendência do IBC-Br com o filtro Boosted HP:

```r
install.packages("tseries")
```

```r
install.packages("expm")
```

```r
download.file("https://raw.githubusercontent.com/zhentaoshi/Boosted_HP_filter/master/R/BoostedHP.R", destfile = "BoostedHP.R")
```

```r
source("BoostedHP.R")
```

```r
bhp = BoostedHP( log( ibc ),
                  lambda = 1600,
                  iter= TRUE,
                  stopping = "BIC", Max_Iter = 100)
```

```r
bhptrend <- bhp$trend
```

Faça um gráfico que contenha duas curvas: o IBC-Br e o exponencial da tendência estimada pelo filtro Boosted HP:

Obtenha o componente cíclico com o filtro Boosted HP:

```r
bhpcycle <- bhp$cycle
```
