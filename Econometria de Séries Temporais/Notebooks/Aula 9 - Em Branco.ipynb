## Teste de Ljung-Box

$\mathcal{H}_0: \sum^{n}_{j=1} \rho_j = 0$ \\
$\mathcal{H}_a: \sum^{n}_{j=1} \rho_j \neq 0$

Vamos realizar o teste de Ljung-Box nas séries da simulação 2:

```r
Box.test( AR, lag = 1, type = "Ljung")
```
