## O modelo

### Sistema de Equações -- 9 variáveis endógenas (c, h, k, w, r, y, i, A e G)

* **Famílias**
  * $c^{-\sigma}_t = \beta E_{t} \left[ c^{-\sigma}_{t+1} \left( 1 + r_{t} \right) \right]$
  * $\frac{w_{t}}{p_{t}}=c_{t}^{\sigma} h_{t}^{\varphi}$
* **Empresas**
  * $y_{t}=A_{t} h_{t}^{1-\alpha}$
  * $\frac{w_{t}}{p_{t}}=(1-\alpha) A_{t} h_{t}^{-\alpha}$
* **Restrição de recursos**
  * $y_t = c_t$
* **Lei de movimento da produtividade**
  * $\ln A_t=(1-\rho_A)\ln\bar{A}+\rho_A\ln A_{t-1}+\varepsilon_t$
* **Política monetária**
  * $R_t = \frac{ 1 }{ \beta } \Pi^{\phi_{\pi}} e^{\epsilon^{m}_t}$
  * $m_t =4 \left( \ln y_t - \ln y_{t-1} \right) - \eta * \left( \ln R_t - \ln R_{t-1} \right) + \ln \Pi_t$
