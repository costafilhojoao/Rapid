## Crie o arquivo .mod

```python
%%writefile ar.mod

var x;

varexo e;

parameters rho,sigma, c;

rho = 0.90;

sigma = 0.01;

c = 1;

model;

x = c + rho * x(-1) + e;

end;

initval;

e = 0;

x = c / ( 1 - rho );

end;

steady;

check;

shocks;

var e;

stderr sigma;

end;

stoch_simul(irf=20, graph_format = pdf);
```

```python
%%writefile solve.m
dynare ar.mod
```
