## O modelo

### Sistema de Equações -- 9 variáveis endógenas (c, h, k, w, r, y, i, A e G)

* **Famílias**
  * $\psi h_t^{ \varphi } c_t^{ \sigma } =  w_t$
  * $c_t^{-\sigma} = \beta E_t\big [ c_{t+1}^{-\sigma} \big( 1 + r_{t+1} - \delta \big) \big]$
  * $k_{t+1}=(1-\delta)k_{t}+i_{t}$
* **Empresas**
  * $y_{t}=A_t k^{\alpha}_{t} h^{1-\alpha}_{t}$
  * $r_t = \alpha \frac{y_{t}}{k_{t}}$
  * $w_t = (1-\alpha) \frac{ y_{t}}{ h_{t}}$
* **Restrição de recursos**
  * $y_t = c_t + i_t + G_t$
* **Lei de movimento da produtividade**
  * $\ln A_t=(1-\rho_A)\ln\bar{A}+\rho_A\ln A_{t-1}+\varepsilon_t$
* **Governo**
  * $\ln G_{t} = (1 - \rho_{G} ) \ln \bar{G}  + \rho_{G} \ln G_{t-1} + \varepsilon_{t}^{G}$
