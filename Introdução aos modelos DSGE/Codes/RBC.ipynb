## Crie o arquivo .mod

```python
%%writefile rbc.mod

var c, h, A, k;

varexo e;

parameters phi, psi, sigma, alpha, beta, delta, rho;

phi   = 1;
psi   = 2.29;
sigma = 2;
alpha = 0.44;
beta  = 0.97;
delta = 0.05;
rho   = 0.9;

model;

# Abar = 1;

psi * exp(h)^phi * exp(c)^sigma = ( 1 - alpha ) * exp(A) * ( exp(k(-1)) / exp(h) )^alpha;

exp(c)^(-sigma) = beta * ( exp(c(+1)) )^(-sigma) * ( 1 + alpha * exp(A(+1)) * ( exp(k) / exp(h(+1)) )^(alpha - 1) - delta );

exp(k) = ( 1 - delta ) * exp(k(-1)) + exp(A) * exp(k(-1))^alpha * exp(h)^( 1 - alpha ) - exp(c);

A = ( 1 - rho ) * Abar + rho * A(-1) + e;

end;

initval;
A = 1;
h = 0.35;
c = 1.01;
k = 9.32;
end;

steady;

check;

shocks;

var e;
stderr 1;
end;

stoch_simul(ar=1, order=1, irf=20);
```

```python
%%writefile solve.m
dynare rbc.mod
```
